```python
import pandas as pd
#geocoder module from google 
import geocoder
import numpy as np
```

```python
data = pd.read_csv("data_sci_snippet.csv")
#create additional column for zipcode 
data["postal"] = np.nan
```

## Fetches zipcode via listing using lat/long and inserts it into empty Postal cell

```python
def find_insert_zipcode(listing):
    try:
        lat = data.GeoLat[listing]
        long = data.GeoLon[listing]
        g = geocoder.google([lat, long], method='reverse')
        post = g.postal
        data.ix[listing,'postal'] = post
    except ValueError:
        post = None
        data.ix[listing,'postal'] = post
```

## Loop through sheet running find_insert_zipcode

```python
for x in range(0,57258):
    find_insert_zipcode(x)
```
